```
#Como sacar las estadisticas de una imgen

def estadisticas(img):

  img_size=img.shape

  if len(img_size)>2:
    print("image chanels", img_size[2])
  else:
    print("image chanels", 1)

  print("image H", img_size[0])
  print("image W", img_size[1])
  print("Min Value", img.min())
  print("Max Value", img.max())
  print("Prom", img.mean())
  print("type", img.dtype)

display_one(img2RGB)
estadisticas(img2RGB)
display_one(imgRGB)
estadisticas(imgRGB)
display_one(imgGRAY)
estadisticas(imgGRAY)
```

```
#Como almacenar y leer multiples imagenes

images = []
IMAGE_DIRECTORY='/content/drive/My Drive/colab/Imagenes'

def get_image(image_path):
  image = cv2.imread(image_path)
  image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
  return image

for file in os.listdir(IMAGE_DIRECTORY):
  #Si no empieza con un punto
  if not file.startswith('.'):
    images.append(get_image(os.path.join(IMAGE_DIRECTORY, file)))

plt.figure(figsize=(20,10))
for i in range (len(images)):
  plt.subplot(1, len(images), i+1)
  plt.imshow(images[i])
  plt.xticks([]), plt.yticks([])
```
