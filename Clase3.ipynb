```
from google.colab import drive
drive.mount('/content/drive')

import numpy as np
import matplotlib.pyplot as plt
import cv2
from google.colab.patches import cv2_imshow

%matplotlib inline

import os
def display1(a, title1 = "Original"):
  plt.imshow(a), plt.title(title1)
  plt.xticks([]), plt.yticks([])
  plt.show()


def display4g(a, b, c, d, title1 = "Original", title2="Edited", title3 = "Edited", title4="Edited"):
  plt.figure(figsize=(10,10))
  plt.subplot(141), plt.imshow(a), plt.title(title1)
  plt.xticks([]), plt.yticks([])
  plt.subplot(142), plt.imshow(b, cmap='gray'), plt.title(title2)
  plt.xticks([]), plt.yticks([])
  plt.subplot(143), plt.imshow(c, cmap='gray'), plt.title(title3)
  plt.xticks([]), plt.yticks([])
  plt.subplot(144), plt.imshow(d, cmap='gray'), plt.title(title4)
  plt.xticks([]), plt.yticks([])
  plt.show()
```

```
#Leo la imagen y creo una figura
imgBGR = cv2.imread('prueba5.jpg', cv2.IMREAD_COLOR)
b = imgBGR[:,:,0]
g = imgBGR[:,:,1]
r = imgBGR[:,:,2]
display4g(imgBGR,b,g,r,"Original","B","G","R")


imgRGB = cv2.cvtColor(imgBGR, cv2.COLOR_BGR2RGB)
r,g,b = cv2.split(imgRGB)
display4g(imgRGB,r,g,b,"Original","R","G","B")
```

```
count = 2

plt.figure(figsize=(12,4))
plt.subplot(1,4,1), plt.imshow(imgRGB/255.0), plt.title("Original RGB")
plt.axis("off")

#i me va a iterar desde cero hasta el tamao del vector y lab va a entregar el valor de cada posicion
for i, lab in enumerate (["R","G","B"]):
  crgb = np.zeros(imgRGB.shape)
  crgb [:,:,i] = imgRGB [:,:,i]
  plt.subplot(1,4,count), plt.imshow(crgb/255.0), plt.title(lab)
  plt.axis("off")
  count+= 1
  

plt.show()
```

```
imgLAB = cv2.cvtColor(imgBGR, cv2.COLOR_BGR2LAB)
L,A,B = cv2.split(imgLAB)
display4g(imgLAB,L,A,B,"LAB","L","A","B")
```
